Now dealing with the missing values so we will replace 37(review rating) missing values with the median of the review but category wise
As the overall median will give bias review rating to the different categories. 

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount($)'})
```

```python
#create column age_group
labels=['Young','Adult','Middle-age','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df['frequency_of_purchases'].unique()
```

```python
#mapping to the values
mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
```

```python
#mapping with the frequency_of_purchase
df['purchase_frequency_days']=df['frequency_of_purchases'].map(mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(21)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
#both columns contain equal values, then we will drop one
df=df.drop('promo_code_used', axis=1)
```

```python
# giving same values to the same no.of days
df["frequency_of_purchases"] = df["frequency_of_purchases"].replace({
    "Bi-Weekly": "Fortnightly",
    "Every 3 Months": "Quarterly"
})
```
